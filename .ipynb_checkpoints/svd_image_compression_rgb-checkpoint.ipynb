```python
import numpy as np
from matplotlib import animation
from matplotlib.image import imread
import matplotlib.pyplot as plt

def im2double(im, min_val, max_val):    
    return (im.astype('float') - min_val) / (max_val - min_val)

img = im2double(imread('mealybugs.jpg'), 0, 255)

plt.figure(figsize=(8, 5))
plt.axis('off')
plt.title('Original')
plt.imshow(img)
plt.show()
```

```python
## Compute SVD
U_red, S_red, VT_red = np.linalg.svd(img[:,:,0], full_matrices=False)
U_green, S_green, VT_green = np.linalg.svd(img[:,:,1], full_matrices=False)
U_blue, S_blue, VT_blue = np.linalg.svd(img[:,:,2], full_matrices=False)

S_red = np.diag(S_red)
S_green = np.diag(S_green)
S_blue = np.diag(S_blue)

plt.figure(figsize=(16,10))

j = 1
for r in (5, 20, 100, 250):
    # Construct approximate image
    img_red_approx = U_red[:,:r] @ S_red[:r,:r] @ VT_red[:r,:]
    img_green_approx = U_green[:,:r] @ S_green[:r,:r] @ VT_green[:r,:]
    img_blue_approx = U_blue[:,:r] @ S_blue[:r,:r] @ VT_blue[:r,:]
    img_approx = np.zeros((img_red_approx.shape[0], img_red_approx.shape[1], 3))
    img_approx[:,:,0] = img_red_approx
    img_approx[:,:,1] = img_green_approx
    img_approx[:,:,2] = img_blue_approx
    img_approx = im2double(img_approx, np.min(img_approx), np.max(img_approx))
    plt.subplot('22'+str(j))
    j += 1
    plt.title('r = '+str(r))
    plt.axis('off')
    plt.imshow(img_approx)

plt.show()
```

```python
# Plot Sigma and Cumulative energy in first 'r' columns/rows
plt.figure(figsize=(16,15))
plt.subplot('321')
plt.title('Singular Values or Sigma Values of Red Channel')
plt.semilogy(np.diag(S_red), color='r')

plt.subplot('322')
plt.title('Cumulative Sum of first r Singular Values of Red Channel')
plt.plot(np.cumsum(np.diag(S_red))/np.sum(np.diag(S_red)), color='r')

plt.subplot('323')
plt.title('Singular Values or Sigma Values of Green Channel')
plt.semilogy(np.diag(S_green), color='g')

plt.subplot('324')
plt.title('Cumulative Sum of first r Singular Values of Green Channel')
plt.plot(np.cumsum(np.diag(S_green))/np.sum(np.diag(S_green)), color='g')

plt.subplot('325')
plt.title('Singular Values or Sigma Values of Blue Channel')
plt.semilogy(np.diag(S_blue), color='b')

plt.subplot('326')
plt.title('Cumulative Sum of first r Singular Values of Blue Channel')
plt.plot(np.cumsum(np.diag(S_blue))/np.sum(np.diag(S_blue)), color='b')

plt.show()
```

```python
# Generate animation of image compression
# Optimal maximum rank value 'r' can be inferred from plot of Sigma values
# Most of the energy of the image is captured within the first 250 vectors
max_r = 250

def truncate_USV(r):
    img_red_approx = U_red[:,:r] @ S_red[:r,:r] @ VT_red[:r,:]
    img_green_approx = U_green[:,:r] @ S_green[:r,:r] @ VT_green[:r,:]
    img_blue_approx = U_blue[:,:r] @ S_blue[:r,:r] @ VT_blue[:r,:]
    img_approx = np.zeros((img_red_approx.shape[0], img_red_approx.shape[1], 3))
    img_approx[:,:,0] = img_red_approx
    img_approx[:,:,1] = img_green_approx
    img_approx[:,:,2] = img_blue_approx
    img_approx = im2double(img_approx, np.min(img_approx), np.max(img_approx))
    
    return img_approx

# Animation of image over point of truncation of SVD
fig, ax = plt.subplots(figsize=(16, 9))
title = ax.set_title('Original Image', fontsize=20)
ax.set_axis_off()
im = ax.imshow(img)
fig.tight_layout()

def init():
    im.set_array(img)
    title.set_text('Original Image')
    return [im]

def animate(i):
    im.set_array(truncate_USV(i))
    title.set_text('Percentage of Data: ' + str(i*100/U_red.shape[1]) + '%')
    return [im]

ani = animation.FuncAnimation(fig, animate, np.arange(max_r, 5, -5), init_func=init, blit=True)
ani.save('img_compress_anim_rgb.mp4', fps=5)
```

```python
# Correlation Matrix
# Interesting to see how different channels correlate with each other
AAT_red = img[:,:,0] @ np.transpose(img[:,:,0])
ATA_red = np.transpose(img[:,:,0]) @ img[:,:,0]

AAT_green = img[:,:,1] @ np.transpose(img[:,:,1])
ATA_green = np.transpose(img[:,:,1]) @ img[:,:,1]

AAT_blue = img[:,:,2] @ np.transpose(img[:,:,2])
ATA_blue = np.transpose(img[:,:,2]) @ img[:,:,2]

plt.figure(figsize=(16,15))
plt.subplot('321')
plt.axis('off')
plt.title('AAT_red')
plt.imshow(AAT_red, cmap='gray')

plt.subplot('322')
plt.axis('off')
plt.title('ATA_red')
plt.imshow(ATA_red, cmap='gray')

plt.subplot('323')
plt.axis('off')
plt.title('AAT_green')
plt.imshow(AAT_green, cmap='gray')

plt.subplot('324')
plt.axis('off')
plt.title('ATA_green')
plt.imshow(ATA_green, cmap='gray')

plt.subplot('325')
plt.axis('off')
plt.title('AAT_blue')
plt.imshow(AAT_blue, cmap='gray')

plt.subplot('326')
plt.axis('off')
plt.title('ATA_blue')
plt.imshow(ATA_blue, cmap='gray')

plt.show()
```
