```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
from matplotlib import animation
import numpy as np
from skimage.color import rgb2lab

def im2double(im, min_val, max_val):    
    return (im.astype('float') - min_val) / (max_val - min_val)

img = im2double(imread('mealybugs.jpg'), 0, 255)

plt.figure(figsize=(8, 5))
plt.axis('off')
plt.title('Original - RGB')
plt.imshow(img)
plt.show()
```

```python
# 0 <= L* <=100
# a* green < 0 < magenta
# b* blue < 0 < yellow

img_lab = rgb2lab(img)

plt.figure(figsize=(8, 12))
plt.subplot('311')
plt.axis('off')
plt.title('L*')
plt.imshow(img_lab[:,:,0], cmap='gray')
plt.subplot('312')
plt.axis('off')
plt.title('a*')
plt.imshow(img_lab[:,:,1], cmap='gray')
plt.subplot('313')
plt.axis('off')
plt.title('b*')
plt.imshow(img_lab[:,:,2], cmap='gray')
plt.show()
```

```python
img_a = img_lab[:,:,1]

img_a_g = np.copy(img_a)
img_a_g[img_a_g > 0] = 0
img_a_g = np.abs(img_a_g)

img_a_m = np.copy(img_a)
img_a_m[img_a_m < 0] = 0
img_a_m = np.abs(img_a_m)

img_b = img_lab[:,:,2]

img_b_b = np.copy(img_b)
img_b_b[img_b_b > 0] = 0
img_b_b = np.abs(img_b_b)

img_b_y = np.copy(img_b)
img_b_y[img_b_y < 0] = 0
img_b_y = np.abs(img_b_y)

plt.figure(figsize=(12, 8))
plt.subplot('221')
plt.axis('off')
plt.title('a* <= 0 (greens)')
plt.imshow(img_a_g, cmap='Greens')
plt.subplot('222')
plt.axis('off')
plt.title('a* >= 0 (magenta)')
plt.imshow(img_a_m, cmap='Reds')
plt.subplot('223')
plt.axis('off')
plt.title('b* <= 0 (blue)')
plt.imshow(img_b_b, cmap='Blues')
plt.subplot('224')
plt.axis('off')
plt.title('b* => 0 (yellow)')
plt.imshow(img_b_y, cmap='binary')
plt.show()
```
